#### Convolutional neural networks (CNN)

Neural networks that analyze data with local patterns — 1D CNNs handle sequential data such as spectra or time series, while 2D CNNs process images such as microscopy or surface maps. They automatically learn filters that detect peaks, edges, or structural features relevant to chemical systems.

#### Recurrent neural networks (RNN)

A type of neural network designed for sequential data, where information flows from one step to the next. In chemistry, RNNs are used to model reaction sequences, analyze time-dependent data, or generate molecular structures from SMILES strings.

[Jupyter notebook 1](ML-RNN.ipynb) *in English*
[Jupyter notebook 2](ML-RNN_2.ipynb) *in English*

#### Autoencoders

Neural networks that compress data into a smaller internal representation (encoding) and then reconstruct it (decoding). In chemistry, they are used to identify hidden structure–property relationships, reduce dimensionality, or generate new molecules with similar features.

[Jupyter notebook](./ML-Autoencoders.ipynb). *in English*
