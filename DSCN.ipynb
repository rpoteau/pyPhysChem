**En résumé** : C'est l'équivalent d'un tableau Excel, mais "mis à nu" sous forme de texte brut pour être universellement compatible.

<div class="intro">

##### Le Coefficient de Corrélation de Pearson
Le coefficient de Pearson ($r$) est l'outil de base pour mesurer la **relation linéaire** entre deux variables numériques (par exemple : est-ce que le pH baisse systématiquement quand l'acidité augmente ?). Il est donné par la formule

$$r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}}$$

* **L'implémentation** : Avec la bibliothèque `pandas`, on calcule l'intégralité de ces relations en une seule ligne de code : `df.corr()`. Cette fonction applique la formule mathématique à chaque paire de colonnes du dataframe.
* **L'échelle** : $r$ varie strictement entre **-1** et **+1**.
* **$r = +1$** : Corrélation positive parfaite (si l'une monte, l'autre monte proportionnellement).
* **$r = -1$** : Corrélation négative parfaite (si l'une monte, l'autre descend).
* **$r = 0$** : Aucune relation linéaire (les deux variables sont indépendantes).

##### La Heatmap (Carte de Chaleur)
C'est la traduction visuelle de la matrice de corrélation. Au lieu de lire un tableau de chiffres indigeste, on utilise des couleurs pour repérer les structures de données.

* **Le principe** : On attribue une **couleur chaude** (rouge/orange) aux corrélations positives fortes et une **couleur froide** (bleu) aux corrélations négatives.
* **L'intérêt en Chimie** : Elle permet de détecter immédiatement les **redondances**. 
    > *Exemple : Si "l'acidité totale" et "l'acidité fixe" ont une corrélation de **0.98**, elles apportent la même information. En garder une seule dans votre modèle XGBoost pourrait suffire à simplifier le problème.*

##### The Pearson Correlation Coefficient

The Pearson coefficient ($r$) is the basic tool for measuring the **linear relationship** between two numerical variables (for example: does the pH systematically decrease when acidity increases?). It is given by the formula
$$r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}}$$
* **Implementation**: With the `pandas` library, all of these relationships are computed in a single line of code: `df.corr()`. This function applies the mathematical formula to each pair of columns in the dataframe.
* **Scale**: $r$ ranges strictly between **-1** and **+1**.
* **$r = +1$**: Perfect positive correlation (if one goes up, the other goes up proportionally).
* **$r = -1$**: Perfect negative correlation (if one goes up, the other goes down).
* **$r = 0$**: No linear relationship (the two variables are independent).

## "Unsupervised learning": Analyse ACP / "Unsupervised learning": PCA analysis

### Rappel de cours : L'Analyse en Composantes Principales (ACP) / Course recap: Principal Component Analysis (PCA)


<div class="intro">

En chimie analytique, nous manipulons souvent des dizaines de variables (pH, taux de sucre, soufre, etc.). C'est ainsi que chaque échantillon de vin $b_i$ est caractérisée par 11 paramètres : $b_i = f(\mathrm{pH}, \mathrm{alcool},...)$. La représentation graphique des bouteilles dans cet espace à 11 dimensions est impossible. 

La PCA est une méthode de **réduction de dimension**. Elle permet de réduire la complexité en créant de nouveaux axes : les **Composantes Principales (PC)**, qui sont décorrélées (orthogonales) entre elles. Cette analyse en composantes principales permet de réduire les observations d’un espace à *n* variables - les descripteurs chimiques - vers un espace à *m* dimensions où *m* est inférieur à *n*. Le but de l’ACP est d’obtenir une représentation plus simple des données sans pour autant perdre de l’information. En général, 2 ou 3 axes suffisent. 

En pratique, on va donc exprimer les coordonnées des bouteilles de vin dans ce nouveau repère, et voir s'il y a un *clustering* des données, c'est-à-dire des sous familles.

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">Pourquoi standardiser les données ?</div>

C'est l'étape la plus critique. En chimie, les unités diffèrent radicalement : le pH varie de 3 à 4, tandis que le soufre total peut monter à 200 mg/L. Pour contourner ce problème, on va utiliser un `StandardScaler`

* Sans standardisation, la PCA accorderait une importance démesurée au soufre simplement parce que ses valeurs numériques sont plus grandes. 
* Le `StandardScaler` transforme chaque variable pour qu'elle ait une **moyenne de 0** et une **variance de 1**. On compare enfin des grandeurs sans dimension.

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">Le principe des Axes Principaux</div>

Contrairement à un graphique classique (ex: pH en fonction de l'Alcool), les axes PC1 et PC2 ne correspondent pas à une seule variable, mais à une **combinaison linéaire** de toutes les variables d'origine. Mathématiquement, ce sont les vecteurs propres de la matrice de covariance (ou de corrélation) des descripteurs.

* **PC1 (Axe de variance maximale)** : PC1 est l'axe qui capture la plus grande part d'information (la variance) du système.
* **PC2 (Axe orthogonal)** : La PCA cherche ensuite la deuxième direction d'étirement, à condition qu'elle soit perpendiculaire à la première. Elle capture l'information secondaire (ex: la structure interne au sein d'une même famille).

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">Interprétation des "Loadings" (Coefficients)</div>

Le tableau des **loadings** indique la contribution (le "poids") de chaque variable d'origine dans la construction des nouveaux axes.
* **Signe positif/négatif** : Si deux variables ont un loading de même signe sur PC1, elles varient ensemble dans le jeu de données. Si les signes sont opposés, elles sont corrélées négativement.
* **Magnitude** : Plus la valeur absolue est proche de 1, plus la variable "pèse" lourd dans la définition de l'axe.

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">Variance expliquée</div>

Lorsqu'on réduit 11 dimensions à seulement 2 (PC1 et PC2), on perd forcément un peu d'information. La variance expliquée mesure quelle proportion du signal d'origine a été conservée dans ce nouveau système. On trouve ici 50%, qui est un très bon score. Cela prouve que le système est multidimensionnel : il n'y a pas qu'un seul facteur qui définit de façon indirecte la couleur d'un vin.

</div>

<div class="introE">
In analytical chemistry, we often work with dozens of variables (pH, sugar level, sulfur, etc.). This is how each wine sample $b_i$ is characterized by 11 parameters: $b_i = f(\mathrm{pH}, \mathrm{alcohol},...)$. Plotting the bottles graphically in this 11-dimensional space is impossible. 

PCA is a **dimensionality reduction** method. It reduces complexity by creating new axes: the **Principal Components (PC)**, which are uncorrelated (orthogonal) with one another. This principal component analysis makes it possible to reduce observations from a space with *n* variables — the chemical descriptors — to a space with *m* dimensions, where *m* is smaller than *n*. The aim of PCA is to obtain a simpler representation of the data without losing information. In general, 2 or 3 axes are enough. 

In practice, we will therefore express the coordinates of the wine bottles in this new reference frame, and see whether there is any *clustering* of the data, i.e. sub-families.

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">Why standardize the data?</div>
This is the most critical step. In chemistry, units differ radically: pH ranges from 3 to 4, whereas total sulfur can go up to 200 mg/L. To get around this problem, we will use a `StandardScaler`
* Without standardization, PCA would give disproportionate importance to sulfur simply because its numerical values are larger. 
* The `StandardScaler` transforms each variable so that it has a **mean of 0** and a **variance of 1**. We are finally comparing dimensionless quantities.

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">The principle of Principal Axes</div>
Unlike a classic plot (e.g. pH as a function of Alcohol), the PC1 and PC2 axes do not correspond to a single variable, but to a **linear combination** of all the original variables. Mathematically, they are the eigenvectors of the covariance (or correlation) matrix of the descriptors.
* **PC1 (Axis of maximum variance)**: PC1 is the axis that captures the largest share of information (the variance) in the system.
* **PC2 (Orthogonal axis)**: PCA then looks for the second direction of spread, provided it is perpendicular to the first. It captures the secondary information (e.g. the internal structure within a single family).

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">Interpreting the "Loadings" (Coefficients)</div>
The table of **loadings** indicates the contribution (the "weight") of each original variable in the construction of the new axes.
* **Positive/negative sign**: If two variables have a loading of the same sign on PC1, they vary together in the dataset. If the signs are opposite, they are negatively correlated.
* **Magnitude**: The closer the absolute value is to 1, the more the variable "weighs" in defining the axis.

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">Explained variance</div>
When we reduce 11 dimensions to just 2 (PC1 and PC2), we inevitably lose a bit of information. The explained variance measures what proportion of the original signal has been retained in this new system. Here we find 50%, which is a very good score. This proves that the system is multidimensional: there is not a single factor that indirectly defines the color of a wine.
</div>

### Arbre de décision / Decision tree

<div class="intro">

Il s'agit d'un algorithme d'apprentissage supervisé, qui divise la population en deux ensembles homogènes ou plus, sur la base des variables indépendantes les plus significatives.

Les arbres de décision ressemblent à des organigrammes, séparant les données en fonction de certaines conditions ou caractéristiques. Ils s'appliquent à la régression aussi bien qu'à la classification.

La structure d'un arbre de décision s'apparente à un organigramme inversé, composé de trois éléments fondamentaux :

- Le Nœud Racine : C'est le point de départ de l'arbre. Il contient l'intégralité du jeu de données et ne possède aucune branche entrante.
- Les Nœuds Internes (ou nœuds de décision) : Alimentés par les branches précédentes, ils évaluent les caractéristiques (variables) des données. Ils effectuent des tests pour diviser la population en sous-ensembles de plus en plus homogènes.
- Les Nœuds Feuilles (ou nœuds terminaux) : Ce sont les extrémités de l'arbre. Ils n'ont aucune branche sortante et représentent la prédiction ou la décision finale de l'algorithme.

Au-delà de sa simplicité, cet algorithme possède deux avantages majeurs : il est extrêmement lisible (ses choix sont transparents et faciles à interpréter pour un être humain) et très rapide à exécuter lors de la phase de prédiction.

  <img src="./ML-Figures/10algos/ArbreDeDecision.jpg" style="margin-left:auto; margin-right:auto; display: block; width:600px" id="img_decision_tree" alt="Exemple d'arbre de décision">
  <figcaption>Exemple schématique d'un arbre de décision</figcaption>
</figure>
</div><br>

<div class="introE">

This is a supervised learning algorithm, which splits the population into two or more homogeneous sets based on the most significant independent variables.

Decision trees resemble flowcharts, separating the data according to certain conditions or features. They apply to regression as well as classification.

The structure of a decision tree is similar to an inverted flowchart, made up of three fundamental elements:
- The Root Node: This is the starting point of the tree. It contains the entire dataset and has no incoming branches.
- The Internal Nodes (or decision nodes): Fed by the preceding branches, they evaluate the features (variables) of the data. They perform tests to split the population into increasingly homogeneous subsets.
- The Leaf Nodes (or terminal nodes): These are the ends of the tree. They have no outgoing branches and represent the algorithm's final prediction or decision.

Beyond its simplicity, this algorithm has two major advantages: it is extremely readable (its choices are transparent and easy for a human to interpret) and very fast to run during the prediction phase.

#### Hyperparamètres / hyperparameters

<div class="intro">

Un arbre de décision laissé sans contraintes grandira jusqu'à ce que chaque feuille soit parfaitement "pure" (Gini = 0). Le modèle finit par apprendre chaque exception individuelle du jeu de données, ce qui mène au **surapprentissage** (overfitting). Pour garantir la capacité de généralisation du modèle, nous ajustons ses **hyperparamètres** (cf la [documentation en ligne](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html)).

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">1. La Profondeur (<code>max_depth</code>)</div>

Il s'agit de la limite du nombre d'étages de l'arbre (le nombre de questions successives).
* **Le principe** : Si on ne fixe pas de limite, l'arbre peut descendre très bas pour isoler chaque bouteille de vin spécifique.
* **L'intérêt** : Une faible profondeur (ex: 3) force l'algorithme à ne capturer que les **règles chimiques majeures** et universelles, ignorant les détails insignifiants.

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">2. Le Seuil de Division (<code>min_samples_split</code>)</div>

C'est le nombre minimum d'échantillons qui doivent être présents dans un nœud pour que l'algorithme soit autorisé à créer une nouvelle branche.
* **Le principe** : Par défaut, ce paramètre est à 2. Si un nœud contient 3 bouteilles, l'algorithme peut encore essayer de les séparer.
* **L'intérêt chimique** : En augmentant ce chiffre (ex: à 50 ou 100), on impose une contrainte statistique : *"Ne crée pas de règle de décision si elle ne concerne pas un groupe significatif de bouteilles"*. Cela évite de créer des branches basées sur des cas isolés ou des erreurs de mesure.

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">3. Le Critère de Qualité (<code>criterion</code>)</div>

C'est la fonction mathématique qui évalue la "pureté" d'une séparation à chaque étape.
* **Gini (par défaut)** : Calcule la probabilité de mauvaise classification. C'est le standard en Data Science car il est robuste et rapide à calculer.
* **Entropy** : Issu de la théorie de l'information, ce critère mesure le "désordre". Bien que mathématiquement différent, il conduit souvent à des résultats très proches du Gini en pratique.

> **💡 Le compromis du Data Scientist :**
> Le réglage de ces paramètres est une quête d'équilibre (le *Bias-Variance Tradeoff*). Si on impose trop de contraintes, l'arbre est trop simple et rate des nuances chimiques (**Sous-apprentissage**). Si on n'en met pas assez, l'arbre devient une "usine à gaz" illisible qui ne fonctionne plus sur de nouvelles données (**Sur-apprentissage**).

</div>

<br>

<div class="ex">

<div class="introE">
A decision tree left unconstrained will grow until every leaf is perfectly "pure" (Gini = 0). The model ends up learning each individual exception in the dataset, which leads to **overfitting**. To ensure the model's ability to generalize, we tune its **hyperparameters** (see the [online documentation](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html)).

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">1. Depth (<code>max_depth</code>)</div>
This is the limit on the number of levels in the tree (the number of successive questions).
* **The principle**: If no limit is set, the tree can go very deep to isolate each specific wine bottle.
* **The value**: A shallow depth (e.g. 3) forces the algorithm to capture only the **major** and universal **chemical rules**, ignoring insignificant details.

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">2. The Split Threshold (<code>min_samples_split</code>)</div>
This is the minimum number of samples that must be present in a node for the algorithm to be allowed to create a new branch.
* **The principle**: By default, this parameter is 2. If a node contains 3 bottles, the algorithm can still try to separate them.
* **The chemical value**: By increasing this number (e.g. to 50 or 100), a statistical constraint is imposed: *"Do not create a decision rule if it does not concern a significant group of bottles"*. This avoids creating branches based on isolated cases or measurement errors.

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">3. The Quality Criterion (<code>criterion</code>)</div>
This is the mathematical function that evaluates the "purity" of a split at each step.
* **Gini (default)**: Computes the probability of misclassification. It is the standard in Data Science because it is robust and fast to compute.
* **Entropy**: Derived from information theory, this criterion measures "disorder". Although mathematically different, it often leads to results very close to Gini in practice.

> **💡 The Data Scientist's tradeoff:**
> Tuning these parameters is a quest for balance (the *Bias-Variance Tradeoff*). If too many constraints are imposed, the tree is too simple and misses chemical nuances (**Underfitting**). If too few are set, the tree becomes an unreadable "gas factory" that no longer works on new data (**Overfitting**).
</div>

### XGBoost  (Gradient Boosting): l'union fait la force / XGBoost (Gradient Boosting): strength in unity

<div class="intro">

**XGBoost** (*eXtreme Gradient Boosting*) est un algorithme d'apprentissage supervisé de type "ensemble". Il est utilisé aussi bien pour des problèmes de **classification** (comme ici pour la couleur du vin) que de **régression**. 

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">1. Le principe de l'Ensemble</div>

Contrairement à l'arbre de décision simple que nous avons vu précédemment, XGBoost ne repose pas sur un seul arbre "expert", mais sur une forêt de centaines de petits arbres. L'idée est de combiner les résultats de nombreux modèles "faibles" pour créer un modèle final extrêmement robuste et précis.

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">2. L'apprentissage séquentiel (Boosting)</div>

XGBoost construit ses arbres de manière **séquentielle** : 
* Chaque nouvel arbre est entraîné spécifiquement pour corriger les erreurs commises par les arbres précédents.
* Le modèle réduit ainsi ses erreurs petit à petit, étape par étape.
* Il utilise une méthode appelée **descente de gradient** : mathématiquement, cela signifie qu'à chaque étape, l'algorithme "descend" dans la direction qui réduit le plus l'erreur globale.

[Image of gradient descent optimization in machine learning]

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">3. Pourquoi est-ce "Extrême" ? (Le complément nécessaire)</div>

Le "X" de XGBoost signifie *eXtreme* car l'algorithme inclut des optimisations avancées :
* **Régularisation L1 et L2** : Il punit mathématiquement les modèles trop complexes pour éviter le surapprentissage (overfitting), bien mieux qu'un arbre de décision standard.
* **Traitement des données manquantes** : Il sait gérer automatiquement les trous dans un jeu de données.
* **Rapidité** : Il est conçu pour le calcul parallèle, ce qui le rend extrêmement rapide même sur des millions de lignes de données.

> **💡 En résumé pour le chimiste :** Si l'arbre de décision classique est une règle à calcul simple et lisible, XGBoost est une station de calcul haute performance. Il perd un peu en "lisibilité immédiate" (on ne peut pas dessiner 500 arbres d'un coup), mais gagne énormément en puissance de prédiction.

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">4. Réintroduire l'eXplicabilité (xAI)</div>
    
Bien que XGBoost soit extrêmement performant, il est beaucoup plus difficile à interpréter qu'un arbre de décision unique : on ne peut pas dessiner et lire 500 arbres à la main. C'est ici qu'intervient l'algorithme **SHAP** (*SHapley Additive exPlanations*).

SHAP est une méthode issue de la théorie des jeux qui permet de "distribuer" la prédiction finale entre tous les descripteurs chimiques. Elle est applicable à tout algorithme de ML, y compris des réseaux de neurones. En ce qui concerne XGBoost, ça permet d'aller polus loin que la quantification de l'importance des descripteurs dans la classification
* Elle nous permet de repasser d'un modèle opaque à une **IA explicable** (xAI).
* Pour chaque bouteille de vin, SHAP mesure exactement comment le pH, le soufre ou l'alcool ont "poussé" le modèle à prédire "Rouge" ou "Blanc".

<div style="color: #004a99; font-weight: bold; font-size: 1em; text-decoration: underline; margin-bottom: 10px;">2. Sequential learning (Boosting)</div>

* Each new tree is trained specifically to correct the errors made by the previous trees.
* The model thus reduces its errors little by little, step by step.
* It uses a method called **gradient descent**: mathematically, this means that at each step, the algorithm "descends" in the direction that most reduces the overall error.

* **L1 and L2 regularization**: It mathematically penalizes overly complex models to avoid overfitting, much better than a standard decision tree.

* **Speed**: It is designed for parallel computing, which makes it extremely fast even on millions of rows of data.

> **💡 In short, for the chemist:** If the classic decision tree is a simple, readable slide rule, XGBoost is a high-performance computing station. It loses a little in "immediate readability" (you can't draw 500 trees at once), but gains enormously in predictive power.

##### Comprendre la "Magie" des Valeurs SHAP (-8 à +8) / Understanding the "Magic" of SHAP Values (-8 to +8)


<div class="hint">

**Indications**

Il peut paraître surprenant que le modèle additionne des valeurs allant de **-8 à +8** pour finalement prédire un simple **0** ou **1**. Voici l'explication scientifique de ce processus :

###### L'Espace Linéaire (Les Log-Odds)
Les valeurs SHAP que nous observons sont exprimées en **Log-Odds** (ou Logit). Dans cet espace :
* **0** représente l'incertitude totale (50% de chance d'être Blanc).
* **+8** est un score extrêmement fort en faveur du **Blanc**.
* **-8** est un score extrêmement fort en faveur du **Rouge**.

Cet espace permet au modèle d'additionner librement l'influence de chaque molécule (Soufre + Alcool + Acidité...) sans être limité par les bornes 0 et 1 à chaque étape du calcul.

###### La Transformation par la Sigmoïde
Pour convertir cette somme mathématique en une probabilité concrète, XGBoost utilise la fonction **Sigmoïde**. Elle compresse les valeurs extrêmes :
* Une somme SHAP de **+5** devient une probabilité de **0,993** (99,3% Blanc).
* Une somme SHAP de **-5** devient une probabilité de **0,006** (99,4% Rouge).
* Une somme SHAP de **0** reste à **0,5** (50/50).



###### La Décision Finale
Enfin, le modèle applique un seuil de décision (par défaut **0,5**) :
* Si $\mathcal{P}(\mathrm{Blanc}) \geq 0,5 \rightarrow$ Classe **1 (Blanc)**.
* Si $\mathcal{P}(\mathrm{Blanc}) < 0,5 \rightarrow$ Classe **0 (Rouge)**.

> **Note pédagogique :** Quand un descripteur comme le `Total Sulfur Dioxide` affiche une valeur SHAP de +4 ou +5, elle a, à elle seule, quasiment "plié le match" pour le modèle, car elle propulse la probabilité au-delà de 98%.

</div><br>
<div class="hintE">

**Hints**

It may seem surprising that the model adds up values ranging from **-8 to +8** to ultimately predict a simple **0** or **1**. Here is the scientific explanation of this process:

###### The Linear Space (Log-Odds)
The SHAP values we observe are expressed in **Log-Odds** (or Logit). In this space:
* **0** represents total uncertainty (50% chance of being White).
* **+8** is an extremely strong score in favor of **White**.
* **-8** is an extremely strong score in favor of **Red**.

This space lets the model freely add up the influence of each molecule (Sulfur + Alcohol + Acidity...) without being limited by the bounds 0 and 1 at each step of the calculation.

###### The Sigmoid Transformation
To convert this mathematical sum into a concrete probability, XGBoost uses the **Sigmoid** function. It compresses the extreme values:
* A SHAP sum of **+5** becomes a probability of **0.993** (99.3% White).
* A SHAP sum of **-5** becomes a probability of **0.006** (99.4% Red).
* A SHAP sum of **0** stays at **0.5** (50/50).

###### The Final Decision
Finally, the model applies a decision threshold (by default **0.5**):
* If $\mathcal{P}(\mathrm{White}) \geq 0.5 \rightarrow$ Class **1 (White)**.
* If $\mathcal{P}(\mathrm{White}) < 0.5 \rightarrow$ Class **0 (Red)**.

> **Pedagogical note:** When a descriptor like `Total Sulfur Dioxide` shows a SHAP value of +4 or +5, it has, on its own, almost "sealed the deal" for the model, because it pushes the probability beyond 98%.

</div>


##### Analyse d'un échantillon de vin / Analysis of a wine sample

<div class="intro">

Après avoir observé le comportement global du modèle avec le *Summary Plot*, nous allons maintenant "zoomer" sur **une seule bouteille de vin**. C'est ce qu'on appelle l'explicabilité locale.

L'objectif de la fonction `plot_prediction_waterfall` est de répondre à la question : *"Exactement pourquoi et comment l'algorithme a-t-il classé cette bouteille spécifique en Rouge ou en Blanc ?"*

La fonction réalise deux opérations majeures, particulièrement adaptées aux chimistes :

1. **La "Traduction" des variables (Inverse Transform)** : L'algorithme a été entraîné sur des données centrées-réduites (ex: $pH = -0.5$). Le code utilise `scaler.inverse_transform` pour réafficher les véritables valeurs chimiques sur le graphique (ex: $pH = 3.2$). C'est indispensable pour garder un sens physico chimique.
2. **La Conversion Mathématique (La Sigmoïde)** : En interne, les arbres de XGBoost ne calculent pas des pourcentages, mais des **Log-odds** (le logarithme de la cote, cote au sens "*bookmaker*" du terme). La fonction additionne ces log-odds, puis utilise une fonction de transformation (la Sigmoïde, ou *expit*) pour ramener ce score abstrait dans un intervalle compris entre 0 et 1 (une vraie probabilité).

***

> **💡 Guide de lecture des deux graphiques générés**
> 
> **1. Le Graphique en Cascade (Waterfall Plot)**
> Il décompose le calcul exact du modèle pour cette bouteille :
> * Le calcul démarre tout en bas à la valeur **$E[f(x)]$ (Base Value)** : c'est la prédiction "moyenne" du modèle si on ne connaissait rien du vin.
> * Chaque paramètre chimique "pousse" ensuite la prédiction vers la gauche (Rouge) ou vers la droite (Blanc).
> * Tout en haut, on obtient le score final **$f(x)$**. C'est la somme exacte de tous les impacts pour cette bouteille précise.
> 
> **2. Le Graphique de Transformation (Sigmoïde)**
> Il illustre la mécanique interne du Machine Learning :
> * Le **point bleu** est le point de départ moyen (la *Base Value* en Log-odds).
> * La **flèche** représente le déplacement total causé par les caractéristiques chimiques uniques de cette bouteille (la somme des valeurs SHAP).
> * Le **point rouge en étoile** est la destination finale. En lisant l'axe vertical (Y), on obtient la probabilité finale que ce vin soit classé "Blanc". Si on dépasse 0.5 (la ligne pointillée noire), le modèle tranche en faveur du blanc !
</div>
<br>
<div class="introE">

After observing the model's overall behavior with the *Summary Plot*, we will now "zoom in" on **a single wine bottle**. This is what is called local explainability.

The goal of the `plot_prediction_waterfall` function is to answer the question: *"Exactly why and how did the algorithm classify this specific bottle as Red or White?"*

The function performs two major operations, particularly well suited to chemists:

1. **The "Translation" of variables (Inverse Transform)**: The algorithm was trained on standardized (centered and scaled) data (e.g. $pH = -0.5$). The code uses `scaler.inverse_transform` to display the real chemical values back on the graph (e.g. $pH = 3.2$). This is essential to keep a physicochemical meaning.
2. **The Mathematical Conversion (The Sigmoid)**: Internally, the XGBoost trees do not compute percentages, but **Log-odds** (the logarithm of the odds, "odds" in the *bookmaker* sense of the term). The function adds up these log-odds, then uses a transformation function (the Sigmoid, or *expit*) to bring this abstract score back into a range between 0 and 1 (a true probability).

***

> **💡 Reading guide for the two generated graphs**
> 
> **1. The Waterfall Plot**
> It breaks down the model's exact calculation for this bottle:
> * The calculation starts at the very bottom at the **$E[f(x)]$ (Base Value)**: this is the model's "average" prediction if we knew nothing about the wine.
> * Each chemical parameter then "pushes" the prediction to the left (Red) or to the right (White).
> * At the very top, we obtain the final score **$f(x)$**. This is the exact sum of all the impacts for this specific bottle.
> 
> **2. The Transformation Graph (Sigmoid)**
> It illustrates the internal mechanics of Machine Learning:
> * The **blue point** is the average starting point (the *Base Value* in Log-odds).
> * The **arrow** represents the total shift caused by the unique chemical characteristics of this bottle (the sum of the SHAP values).
> * The **red star point** is the final destination. By reading the vertical (Y) axis, we obtain the final probability that this wine is classified as "White". If we exceed 0.5 (the black dotted line), the model decides in favor of white!
</div>

##### Code python / Python code

````python
import shap
import matplotlib.pyplot as plt
import scipy

def plot_prediction_waterfall(model, X_scaled, y, y_bin, feature_names, scaler, sample_index=0):
    """
    Creates a waterfall plot for a single prediction.
    Shows the journey from the base value to the final log-odds.
    """
    # 1. Setup Explainer
    explainer = shap.TreeExplainer(model)
    shap_values_4waterFall = explainer(X_scaled)
    shap_values = explainer.shap_values(X_scaled)
    
    # 2. Get the specific sample's explanation
    sample_explanation = shap_values_4waterFall[sample_index]
    
    # 3. Replace the scaled values with real values for the display
    # This ensures the plot shows "pH = 3.2" instead of "pH = -0.5"
    X_original = scaler.inverse_transform(X_scaled)
    sample_explanation.data = X_original[sample_index]
    sample_explanation.feature_names = list(feature_names)

    # 4. Plot
    plt.figure(figsize=(10, 6))
    shap.plots.waterfall(sample_explanation)
    plt.show()

    print("-" * 65)
    base_value = explainer.expected_value
    sample_explanation = shap_values[sample_index]
    total_log_odds = base_value + np.sum(sample_explanation)
    probability = scipy.special.expit(total_log_odds)
    print(f"Base value         : {base_value:.4f}")
    print(f"Sum of SHAP Values : {np.sum(sample_explanation):.4f}")
    print(f"TOTAL LOG-ODDS     : {total_log_odds:.4f}")
    print(f"FINAL PROBA        : {probability:.4%}")
    print(f"Real Probability   : {100*y_bin.iloc[sample_index]}% ({y.iloc[sample_index]})")
    print("-" * 65)

    plt.figure(figsize=(10, 5))
    
    # Création de la courbe sigmoïde
    x_range = np.linspace(-10, 10, 400)
    y_range = scipy.special.expit(x_range)
    plt.plot(x_range, y_range, color="#1f8b91", label='Sigmoid Function')
    
    # Ligne de décision à 0.5
    plt.axhline(0.5, color='black', linestyle='--', alpha=0.3)
    plt.axvline(0, color='black', linestyle='--', alpha=0.3)

    # Point de départ (Base Value)
    plt.scatter(base_value, scipy.special.expit(base_value), color='blue', s=100, label='Base Value (Start)', zorder=5)
    
    # Point final (Prediction)
    plt.scatter(total_log_odds, probability, color='red', s=150, marker='*', label='Final Prediction', zorder=6)

    # Flèche montrant le déplacement
    plt.annotate('', xy=(total_log_odds, probability), xytext=(base_value, scipy.special.expit(base_value)),
                 arrowprops=dict(facecolor='black', color='grey', shrink=0.05, width=1, headwidth=8))

    plt.title(f"Logistic Transformation (Log-odds to Probability)\nSample {sample_index}")
    plt.xlabel("Total Log-odds (Sum of SHAP + Base Value)", fontsize=12)
    plt.ylabel("Probability", fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.2)
    plt.show()
````
