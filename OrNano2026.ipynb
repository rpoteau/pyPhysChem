### Qu’est-ce qu’une API ?

Une **API** (*Application Programming Interface*, interface de programmation applicative) est un ensemble de règles et de fonctions qui permet à deux programmes de communiquer.  

👉 Analogie : c’est comme le **menu d’un restaurant**  
- Vous (l’utilisateur) choisissez un plat → **la requête**  
- La cuisine (le serveur) le prépare → **le traitement**  
- Le serveur ou le serveur de salle (l’API) vous apporte le plat → **la réponse**  

Vous n’avez pas besoin de savoir comment la cuisine fonctionne ni d'être inventif en cuisine, seulement comment lire le menu (la documentation de l’API).

### Lecture des données et stockage dans un dataframe pandas

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from IPython.display import display

def readWines(prt=True, merge=False):
    from pathlib import Path
    
    pathway = Path() / "ML-data"
    dataRedWineFile=pathway / 'winequality-red.csv'
    dfRedWine=pd.read_csv(dataRedWineFile,sep=";",header=0)
    dfRedWine["color"] = "red" 
    if (prt): display(dfRedWine)
    # describe() generates descriptive statistics
    if (prt): display(dfRedWine.describe().style.format("{0:.2f}"))
    if (prt): print(dfRedWine.groupby('quality').size())
    
    dataWhiteWineFile=pathway / 'winequality-white.csv'
    dfWhiteWine=pd.read_csv(dataWhiteWineFile,sep=";",header=0)
    dfWhiteWine["color"] = "white" 
    if (prt): display(dfWhiteWine)
    # describe() generates descriptive statistics
    if (prt): display(dfWhiteWine.describe().style.format("{0:.2f}"))
    if (prt): print(dfWhiteWine.groupby('quality').size())

    if merge:
        dfRWWines = pd.concat([dfRedWine,dfWhiteWine],axis=0)
        # drop=True for not adding the index column to the df
        # inplace=True for updating df inplace without copying to another object. 
        dfRWWines.reset_index(drop=True, inplace=True)
        return dfRWWines
    else:
        return dfRedWine,dfWhiteWine

dfRedWine, dfWhiteWine = readWines(prt=False, merge=False)

display(dfRedWine)
display(dfWhiteWine)
````

## "Unsupervised learning": Analyse PCA

### Le code python

````python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Applique le style fond blanc avec grille grise discrète
plt.style.use('seaborn-v0_8-white')

# 1. Préparation des données (on ne garde que les colonnes numériques)
x = dfRWWines[features].values
y = dfRWWines['color'] # Pour la coloration finale
y_bin = dfRWWines['color_bin'] # Pour XGBoost

# 2. Standardisation (Étape cruciale en chimie !)
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# 3. Calcul de la PCA (2 composantes pour le graphique)
pca = PCA(n_components=2)
pca_result = pca.fit_transform(x_scaled)


# --- VISUALISATION ---
fig, ax = plt.subplots(1, 1, figsize=(8, 8), constrained_layout=True)

# Plot PCA
sns.scatterplot(x=pca_result[:,0], y=pca_result[:,1], hue=y, palette=['red', 'gold'], alpha=0.5, ax=ax)
ax.set_title(f'PCA - Variance expliquée: {pca.explained_variance_ratio_.sum():.2%}')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
plt.show()
# Afficher l'influence des variables sur PC1 et PC2
loadings = pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2'], index=features)
print("Influence des variables sur les axes :")
display(loadings.sort_values(by='PC1', ascending=False))
````

### 🍷 Comment interpréter cette projection ? Est-ce que ça fait sens &oelig;nologique ?

#### L'axe 1 (PC1) est le marqueur de la couleur, puisqu'il y a séparation rouge/blanc selon cet axe

- **Côté Positif. Descripteurs typiques des vins Blancs**
    - Total SO2 (+0.49) et Free SO2 (+0.43). les vins blancs ne bénéficient pas des tanins (polyphénols) des peaux de raisin, qui sont des antioxydants naturels. Pour éviter qu'un vin blanc ne brunisse ou ne s'oxyde, l'œnologue est obligé d'ajouter plus de sulfites
    - Sucre résiduel (+0.34) : on le retrouve souvent associé aux blancs dans ces jeux de données, car les vins liquoreux ou demi-secs sont quasi exclusivement des blancs
- **Côté Négatif. Descripteurs typiques des vins Rouges**
    - Volatile Acidity (-0.38) : les rouges subissent souvent des macérations plus longues et une fermentation malolactique systématique. Ces processus augmentent naturellement l'acidité volatile (essentiellement l'acide acétique) par rapport aux blancs, qui sont filtrés plus rapidement
    - Chlorides (-0.29) : témoignent de l'extraction minérale plus intense lors de la macération des peaux, spécifique au rouge
    - Sulphates (-0.29) : attention à ne pas confondre avec les sulfites (SO₂). Les sulfates dans le rouge proviennent souvent de l'extraction des matières solides (peaux, pépins) et parfois de l'usage de sulfate de cuivre à la vigne
    - Fixed Acidity (-0.24) : l'acide tartrique est souvent plus présent dans les vins rouges.

#### L'axe 2 est dominé par deux variables opposées : Density et Alcohol

- **Côté Positif. Le "Lourd"**.  Density (+0.58), Residual Sugar (+0.33), Fixed Acidity (+0.34)
    Plus un vin monte sur l'axe 2, plus il est "dense" et chargé en extraits secs. On y trouve des vins qui ont soit beaucoup de sucre résiduel, soit une forte acidité fixe, ce qui augmente la densité physique du liquide.

- **Côté Négatif. Le "Léger/Alcooleux"**. Alcohol (-0.46)
    C'est l'opposé mathématique de la densité (l'éthanol est moins dense que l'eau). Un vin situé en bas de l'axe 2 est un vin avec un fort degré alcoolique et moins de corps/sucre

## Supervised learning

### Code pour tracer des matrices de confusion à la fin

````python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

plt.style.use('seaborn-v0_8-white')

def plot_dual_confusion_matrices(y_train_true, y_train_pred, y_test_true, y_test_pred, title="Model"):
    """
    Plots side-by-side confusion matrices for training and testing sets.
    
    Args:
        y_train_true, y_train_pred: True and predicted labels for the training set.
        y_test_true, y_test_pred: True and predicted labels for the test set.
        title: Main title for the figure (e.g., "XGBoost" or "Decision Tree").
    """
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    

    # 1. Training Confusion Matrix
    cm_train = confusion_matrix(y_train_true, y_train_pred)
    sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues', 
                xticklabels=labels, yticklabels=labels, ax=axes[0])
    axes[0].set_title(f'{title} - Training Set')
    axes[0].set_xlabel('Predicted')
    axes[0].set_ylabel('Actual')

    # 2. Testing Confusion Matrix
    cm_test = confusion_matrix(y_test_true, y_test_pred)
    sns.heatmap(cm_test, annot=True, fmt='d', cmap='Greens', 
                xticklabels=labels, yticklabels=labels, ax=axes[1])
    axes[1].set_title(f'{title} - Test Set')
    axes[1].set_xlabel('Predicted')
    axes[1].set_ylabel('Actual')

    plt.tight_layout()
    plt.show()
````

### Partition des données pour l'apprentissage et la validation

````python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 1. Préparation des données (on ne garde que les colonnes numériques)
x = dfRWWines[features].values
y = dfRWWines['color'] # Pour la coloration finale
y_bin = dfRWWines['color_bin'] # Pour XGBoost

# 2. Standardisation (Étape cruciale en chimie !)
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Définition des variables explicatives (X) et de la cible (y)
# D'après votre notebook, y contient la couleur (0 pour rouge, 1 pour blanc par exemple)
X = x_scaled
# y est déjà défini plus haut dans votre notebook

# Partition 80% entraînement / 20% test
# random_state=42 permet de figer le hasard pour avoir toujours les mêmes résultats
# l'arbre de decision utilisera X_train, X_test, y_train, y_test
# XGBoost utilisera X_train, X_test,  y_bin_train, y_bin_test
X_train, X_test, y_train, y_test, y_bin_train, y_bin_test = train_test_split(X, y, y_bin, test_size=0.20, random_state=42)

print(f"Nombre d'échantillons d'entraînement : {X_train.shape[0]}")
print(f"Nombre d'échantillons de test : {X_test.shape[0]}")
````

### Decision tree

#### Le code python

````python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Création et entraînement du modèle
# On limite la profondeur (max_depth) pour que l'arbre reste lisible
dt_model = DecisionTreeClassifier(max_depth=3, random_state=42)
dt_model.fit(X_train, y_train)

# 2. Prédictions
y_test_hat_dt = dt_model.predict(X_test)
y_train_hat_dt = dt_model.predict(X_train)

# 3. Affichage de la précision
print(f"Précision Arbre de Décision sur l'ensemble d'entraînement: {accuracy_score(y_train, y_train_hat_dt):.2%}")
print(f"Précision Arbre de Décision sur l'ensemble de test       : {accuracy_score(y_test, y_test_hat_dt):.2%}")
````

#### Matrices de confusion

````python
plot_dual_confusion_matrices(
    y_train, y_train_hat_dt, 
    y_test, y_test_hat_dt, 
    title="Confusion de l'arbre de décision"
)
````

#### Comment interpréter l'arbre ?

**Chaque "boîte" (nœud) de l'arbre contient des informations clés pour la classification**

1. **condition** (ex: `chlorides <= -0.272`) = critère chimique le plus discriminant à cette étape. Si le vin remplit la condition, on va à gauche, sinon à droite
2.  **gini** = indice d'impureté. `0.0` signifie que le groupe est "pur" (tous les vins sont de la même couleur). Plus il est proche de `0.5`, plus le mélange est équilibré entre rouge et blanc
3.  **samples** = nombre de bouteilles de vin analysées dans ce nœud
4.  **value** = répartition exacte. `[987, 95]` signifie 987 vins rouges et 95 vins blancs
5.  **class** = couleur majoritaire qui serait prédite si l'on s'arrêtait là

####  🍷 Analyse de l'Arbre de Décision (gemini, 2 mars 2026)

> **Note technique :** Les seuils affichés (ex: -0.89) correspondent à des données **centrées-réduites** (`StandardScaler()`). 
> * `0` représente la moyenne du jeu de données.
> * Une valeur négative est inférieure à la moyenne.

**1. Le décideur majeur : Le Dioxyde de Soufre (`total sulfur dioxide`)**
La racine de l'arbre (le premier test) repose sur le taux de SO<sub>2</sub>. 
* **Observation :** Si `total sulfur dioxide <= -0.89`, le modèle s'oriente vers le vin **rouge**.
* **Interprétation :** Cela confirme la littérature œnologique : les vins blancs nécessitent généralement des doses de sulfites plus élevées pour leur stabilisation et protection contre l'oxydation par rapport aux vins rouges.

**2. Le second critère : Les Chlorures (`chlorides`)**
Au deuxième niveau, l'algorithme utilise la concentration en sels.
* **Observation :** Un taux de soufre bas couplé à des chlorures élevés (`> -0.27`) mène quasi systématiquement à une classification en **vin rouge**.
* **Interprétation :** Les chlorures sont souvent liés au terroir et aux pratiques de vinification ; ils constituent ici une signature chimique discriminante.

**3. Zones d'incertitude**
L'arbre montre une branche plus complexe quand le soufre et les chlorures sont tous deux très bas (`total sulfur dioxide <= -0.89` et `chlorides <= -0.27`).
* **Observation :** L'algorithme doit alors effectuer un troisième test sur le soufre.
* **Interprétation :** Cela révèle une zone de recouvrement physico-chimique où les profils des vins blancs et rouges légers sont très proches.

**4. Le "juge de paix" : L'Acidité Volatile (`volatile acidity`)**
Ce critère n'intervient qu'en fin de parcours pour les échantillons les plus ambigus.
* **Interprétation :** L'acidité volatile est typiquement plus élevée dans les vins rouges du fait des processus de fermentation et d'élevage. L'arbre l'utilise pour affiner la séparation finale.

#### Hyperparamètres

Pour ajuster l'arbre (et éviter l'overfitting), vous pouvez modifier ces arguments dans `DecisionTreeClassifier()` :

- `max_depth` : La profondeur maximale. Plus elle est grande, plus l'arbre est complexe, mais plus le risque d'overfitting est élevé.
- `min_samples_split` : Le nombre minimum d'échantillons requis pour diviser un nœud. Augmenter ce chiffre simplifie l'arbre.
- `criterion` : `gini` (par défaut) ou `entropy`. Ce sont deux manières mathématiques de calculer la "pureté" des groupes.

#### Comprendre la "Magie" des Valeurs SHAP (-8 à +8)

Il peut paraître surprenant que le modèle additionne des valeurs allant de **-8 à +8** pour finalement prédire un simple **0** ou **1**. Voici l'explication scientifique de ce processus :

##### L'Espace Linéaire (Les Log-Odds)
Les valeurs SHAP que nous observons sont exprimées en **Log-Odds** (ou Logit). Dans cet espace :
* **0** représente l'incertitude totale (50% de chance d'être Blanc).
* **+8** est un score extrêmement fort en faveur du **Blanc**.
* **-8** est un score extrêmement fort en faveur du **Rouge**.

Cet espace permet au modèle d'additionner librement l'influence de chaque molécule (Soufre + Alcool + Acidité...) sans être limité par les bornes 0 et 1 à chaque étape du calcul.

##### La Transformation par la Sigmoïde
Pour convertir cette somme mathématique en une probabilité concrète, XGBoost utilise la fonction **Sigmoïde**. Elle compresse les valeurs extrêmes :
* Une somme SHAP de **+5** devient une probabilité de **0,993** (99,3% Blanc).
* Une somme SHAP de **-5** devient une probabilité de **0,006** (99,4% Rouge).
* Une somme SHAP de **0** reste à **0,5** (50/50).



##### La Décision Finale
Enfin, le modèle applique un seuil de décision (par défaut **0,5**) :
* Si $\mathcal{P}(\mathrm{Blanc}) \geq 0,5 \rightarrow$ Classe **1 (Blanc)**.
* Si $\mathcal{P}(\mathrm{Blanc}) < 0,5 \rightarrow$ Classe **0 (Rouge)**.

> **Note pédagogique :** Quand un descripteur comme le `Total Sulfur Dioxide` affiche une valeur SHAP de +4 ou +5, elle a, à elle seule, quasiment "plié le match" pour le modèle, car elle propulse la probabilité au-delà de 98%.
