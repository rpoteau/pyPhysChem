### R√©solution <u>num√©rique</u> d'√©quation cin√©tique coupl√©es = syst√®me d'EDO

\begin{equation}
    A \xrightarrow{k_1} B
\end{equation}

\begin{equation}
    B \xrightarrow{k_2} C
\end{equation}

expressions des lois de vitesse :

$\frac{d c_A}{dt} = -k_1 c_A$

$\frac{d c_B}{dt} = k_1 c_A - k_2 c_B$

$\frac{d c_C}{dt} = k_2 c_B$

````python
import numpy as np

def dF(Y,t):
    val1 = -k1*Y[0]
    val2 =  k1*Y[0] - k2*Y[1]
    val3 =  k2*Y[1]
    return val1, val2, val3

k1 = 0.005 #s-1
k2 = 0.005 #s-1
Y0 = np.array([1e-3,0,0]) #mol.L-1
dY0 = dF(Y0,0)

print("(d[A]/dt)(t=0)=",dY0[0],'mol.L-1.s-1')
print("(d[B]/dt)(t=0)=",dY0[1],'mol.L-1.s-1')
print("(d[C]/dt)(t=0)=",dY0[2],'mol.L-1.s-1')
````

````python
from scipy.integrate import odeint
from matplotlib import pyplot as plt

t = np.linspace(0,2000,100) #s
sol = odeint(dF,Y0,t)

ca = sol[:,0]
cb = sol[:,1]
cc = sol[:,2]

i = np.argmax(cb)
t_max = t[i]
cb_max = cb[i]
print(f"tmax = {t_max:.1f} s")
print(f"cb_max = {cb_max:.5f} mol.L-1")

plt.figure()
plt.plot(t,sol,label=["[A]","[B]","[C]"],linewidth=3)
plt.plot([t_max, t_max],[0,cb_max],ls='--',c='r',lw=1)
plt.plot([0, t_max],[cb_max, cb_max],ls='--',c='r',lw=1)
plt.plot(t,ca+cb+cc,label="[A]+[B]+[C]",ls='-.',lw=1)
plt.xlabel('t / s')
plt.ylabel('c / mol.L-1')
plt.xlim(0,2000)
plt.ylim(0,)
plt.legend()
````

### R√©solution <u>analytique</u> d'√©quations math√©matiques

<figure> 
<img width = "800px" src="./MiscImages/logos/SympyLogo.svg" alt="logoSympy" />

**Exemple: Cin√©tique d'ordre un d'une r√©action chimique**

Le syst√®me √† r√©soudre est :

$$\left\{ \begin{array}{c}
-\frac{d[\mathrm{A}]_t}{dt} - k[\mathrm{A}]_t = 0\\
[\mathrm{A}]_{t=0}=a_{0}
\end{array}\right.$$

````python
import sympy as sym
from tools4pyPhysChem import PrintLatexStyleSymPyEquation as ple

# d√©claration des symboles des variables et des fonctions
t = sym.Symbol('t')
k = sym.Symbol('k')
A = sym.Function('A')
# d√©finition des fonctions
dAdt = A(t).diff(t)
expr = sym.Eq(dAdt+k*A(t), 0)

ple(dAdt)
ple(expr)

sym.dsolve(expr,A(t),ics={A(0): 'a0'})
````

### Visualisation d'une grande BDD de mol√©cules via leur similarit√© chimique

````python
import pandas as pd
import io

t4pPC.centerTitle("Lecture d'un fichier excel contenant des SMILES")
df = pd.read_excel("./ML-data/BBBP_nettoye.xlsx")

df.head()
````

````python
from rdkit.Chem import PandasTools

from rdkit import RDLogger
# D√©sactive les messages de niveau WARNING et inf√©rieurs
lg = RDLogger.logger()
lg.setLevel(RDLogger.ERROR)

def smiles_to_mol_validator(smiles, prt=False):
    """
    Convertit un SMILES, nettoie les fragments monoatomiques,
    et retourne √† la fois l'objet Mol ET le SMILES canonique nettoy√©.
    retourne None, None en cas de probl√®me de conversion
    """
    try:
        # Tente de cr√©er l'objet Mol
        mol = Chem.MolFromSmiles(smiles)
        
        # Le MolFromSmiles de RDKit retourne None si le SMILES est invalide.
        # On s'assure qu'il s'agit bien d'une structure (non None et ayant des atomes)
        if mol is not None and mol.GetNumAtoms() > 0:
            frags = Chem.GetMolFrags(mol, asMols=True)
            # D√©tection d'un H isol√©
            is_problematic = any(f.GetNumAtoms() == 1 for f in frags)
            if is_problematic: 
                if prt: display(mol)
                # On ne garde que les fragments "complexes" (plus de 1 atome)
                # Cela supprimera H+, Cl-, Na+, etc.
                clean_frags = [f for f in frags if f.GetNumAtoms() > 1]
                if clean_frags:
                    mol = clean_frags[0]
                    for i in range(1, len(clean_frags)):
                        mol = Chem.CombineMols(mol, clean_frags[i])
                    Chem.SanitizeMol(mol)
                    if prt: display(mol)
                else:
                    return None
            # On g√©n√®re le nouveau SMILES propre (canonique)
            clean_smiles = Chem.MolToSmiles(mol)
            return clean_smiles
        else:
            return None
    except Exception:
        return None

t4pPC.centerTitle("Curation des donn√©es & visualisation 2D des mol√©cules")
df['clean_smiles'] = df['smiles'].apply(lambda x: pd.Series(smiles_to_mol_validator(x)))
PandasTools.AddMoleculeColumnToFrame(df, smilesCol='clean_smiles',includeFingerprints=True,molCol='mol') #inserted as the last column
display(df)
````

````python
def check_lipinski_veber(row):

    violations = 0
    
    # Crit√®res de Lipinski
    if row['MW'] > 500: violations += 1
    if row['LogP'] > 5: violations += 1
    if row['HBD'] > 5: violations += 1
    if row['HBA'] > 10: violations += 1

    # Crit√®res de Veber
        # 1. TPSA <= 140 √Ö¬≤
        # 2. Liaisons "rotatives" <= 10
    veber_violation = False
    if row['TPSA'] > 140 or row['NumRotatableBonds'] > 10:
        veber_violation = True
    
    lipinski_pass = 1 if violations <= 1 else 0
    
    return pd.Series({
        'Lipinski_Pass': lipinski_pass,
        'Violations_Count': violations,
        'Veber_Pass': 0 if veber_violation else 1
    })
````

````python
def count_sp3_carbons(mol):
    """Compte le nombre de carbones ayant une hybridation sp3."""
    count = 0
    for atom in mol.GetAtoms():
        if atom.GetSymbol() == 'C' and atom.GetHybridization() == Chem.rdchem.HybridizationType.SP3:
            count += 1
    return count
    
def compute_all_descriptors(mol):
    from rdkit.Chem import Descriptors
    from rdkit.Chem import Lipinski
    from rdkit.Chem import rdMolDescriptors
    """Calcule un ensemble de descripteurs physico-chimiques et structurels RDKit."""
    
    # Calcul des descripteurs
    mw = Descriptors.MolWt(mol)
    logp = Descriptors.MolLogP(mol)
    hbd = Lipinski.NumHDonors(mol)
    hba = Lipinski.NumHAcceptors(mol)
    tpsa = Descriptors.TPSA(mol)
    rotatable_bonds = Descriptors.NumRotatableBonds(mol)
    num_aromatic_rings = Descriptors.NumAromaticRings(mol)
    num_cycles = Descriptors.RingCount(mol)
    qed = Descriptors.qed(mol)
    sp3_carbons = count_sp3_carbons(mol)
    
    return pd.Series({
        'MW': mw,
        'LogP': logp,
        'HBD': hbd,
        'HBA': hba,
        'TPSA': tpsa,
        'NumRotatableBonds': rotatable_bonds,
        'NumAromaticRings': num_aromatic_rings,
        'NumSP3Carbons': sp3_carbons,
        'NumCycles': num_cycles,
        'QED': qed
    })

t4pPC.centerTitle("Calcul des descripteurs")

# Appliquer la fonction de calcul des descripteurs sur la colonne 'mol'
descriptor_df = df['mol'].apply(compute_all_descriptors)

# Fusionner les descripteurs avec le DataFrame nettoy√©, et supprimer la colonne 'mol'
df_final = pd.concat([df.drop(columns=['mol']), descriptor_df], axis=1)
print("Done")
````

````python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np

# R√©cup√©ration des variables du contexte pr√©c√©dent
# Assurez-vous que ces variables sont d√©finies si vous ex√©cutez ce bloc seul
# Pour cet exemple, je vais les initialiser comme si elles venaient du script pr√©c√©dent
file_path = './MiscData/exponentialLawInANarrowDomain.xlsx'
try:
    df = pd.read_excel(file_path)
except FileNotFoundError:
    print(f"Erreur : Le fichier '{file_path}' n'a pas √©t√© trouv√©. Veuillez v√©rifier le chemin d'acc√®s.")
    exit()
except Exception as e:
    print(f"Une erreur est survenue lors du chargement du fichier Excel : {e}")
    exit()

time_column = 't / s'
concentration_column = 'C / mol.L-1'

# V√©rifier si les colonnes existent (re-v√©rification pour la robustesse)
if time_column not in df.columns or concentration_column not in df.columns:
    print(f"\nErreur : Les colonnes '{time_column}' ou '{concentration_column}' n'ont pas √©t√© trouv√©es dans le fichier Excel.")
    print("Colonnes disponibles :", df.columns.tolist())
    print("Veuillez ajuster les variables 'time_column' et 'concentration_column' dans le code.")
    exit()

# --- √âtape 1 : V√©rifier et transformer les donn√©es pour la loi exponentielle ---
# S'assurer que les valeurs de concentration sont positives pour prendre le log
if (df[concentration_column] <= 0).any():
    print(f"\nAttention : La colonne '{concentration_column}' contient des valeurs non positives (z√©ro ou n√©gatives).")
    print("Impossible d'appliquer le logarithme naturel. Veuillez v√©rifier vos donn√©es.")
    exit()

df['log_Concentration'] = np.log(df[concentration_column])
print(f"\nLa colonne 'log_Concentration' a √©t√© cr√©√©e en prenant le logarithme naturel de '{concentration_column}'.")
print("\nVoici les 5 premi√®res lignes du DataFrame avec la nouvelle colonne :")
print(df.head())

# --- √âtape 2 : Effectuer une r√©gression lin√©aire sur les donn√©es transform√©es ---
x_exp = df[time_column]
y_exp_transformed = df['log_Concentration']

slope_exp, intercept_exp, r_value_exp, p_value_exp, std_err_exp = linregress(x_exp, y_exp_transformed)

print(f"\nR√©sultats de la r√©gression lin√©aire sur ln(Concentration) vs Temps :")
print(f"  Pente (B) : {slope_exp:.4f}")
print(f"  Ordonn√©e √† l'origine (ln(A)) : {intercept_exp:.4f}")
print(f"  Coefficient de corr√©lation (R) : {r_value_exp:.4f}")
print(f"  Coefficient de d√©termination (R¬≤) : {r_value_exp**2:.4f}")
print(f"  Valeur p : {p_value_exp:.4f}")
print(f"  Erreur standard de la pente : {std_err_exp:.4f}")

# Reconstruire les param√®tres A et B pour le mod√®le exponentiel original
# B = slope_exp
# A = exp(intercept_exp)
A_exp = np.exp(intercept_exp)
B_exp = slope_exp

print(f"\nParam√®tres du mod√®le exponentiel Y = A * exp(B * X) :")
print(f"  A (facteur pr√©-exponentiel) : {A_exp:.4f}")
print(f"  B (taux exponentiel) : {B_exp:.4f}")
print(f"L'√©quation exponentielle ajust√©e est donc : Concentration = {A_exp:.4f} * exp({B_exp:.4f} * Temps)")


# --- √âtape 3 & 4 : Tracer la r√©gression lin√©aris√©e et ses r√©sidus ---
y_exp_pred_transformed = slope_exp * x_exp + intercept_exp
residuals_exp = y_exp_transformed - y_exp_pred_transformed

plt.figure(figsize=(12, 6))

# Subplot 1: log(Concentration) vs Time with linear regression
plt.subplot(1, 2, 1) # 1 ligne, 2 colonnes, premier plot
plt.plot(x_exp, y_exp_transformed, 'o', label='Donn√©es transform√©es (ln(Concentration))')
plt.plot(x_exp, y_exp_pred_transformed, 'r-', label=f'R√©gression lin√©aire: ln(Y) = {slope_exp:.2f}X + {intercept_exp:.2f} (R¬≤ = {r_value_exp**2:.2f})')
plt.title(f'ln(Concentration) vs Temps avec r√©gression lin√©aire ({file_path.split("/")[-1]})')
plt.xlabel(f'{time_column} (unit√© de temps)')
plt.ylabel(f'ln({concentration_column})')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Subplot 2: R√©sidus de la r√©gression lin√©aris√©e
plt.subplot(1, 2, 2) # 1 ligne, 2 colonnes, deuxi√®me plot
plt.plot(x_exp, residuals_exp, 'o', color='green', label='R√©sidus (mod√®le exponentiel lin√©aris√©)')
plt.axhline(0, color='gray', linestyle='--', linewidth=1, label='R√©sidu = 0')
plt.title(f'R√©sidus de la r√©gression de ln(Concentration) vs Temps')
plt.xlabel(f'{time_column} (unit√© de temps)')
plt.ylabel('R√©sidus (ln(Y) observ√© - ln(Y) pr√©dit)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

plt.tight_layout() # Ajuste automatiquement les param√®tres des sous-graphiques pour qu'ils s'adaptent √† la zone de figure
plt.show()

print(f"\nLes graphiques de ln(Concentration) en fonction du temps et de ses r√©sidus ont √©t√© g√©n√©r√©s.")
print(f"Si les points de ln(Concentration) forment une ligne droite et que les r√©sidus sont al√©atoires autour de z√©ro,")
print(f"cela indique que le mod√®le exponentiel est une bonne description de vos donn√©es.")


# --- √âtape 6 : Comparer les mod√®les sur le graphique original ---
# Nous allons re-calculer les pr√©dictions du mod√®le lin√©aire original pour la comparaison
# (Ces valeurs ont d√©j√† √©t√© calcul√©es dans la r√©ponse pr√©c√©dente, mais les recr√©ons pour l'ind√©pendance)
x_original = df[time_column]
y_original = df[concentration_column]

# Mod√®le lin√©aire (du code pr√©c√©dent)
# Pour une comparaison juste, je vais re-ex√©cuter la r√©gression lin√©aire simple ici
slope_lin, intercept_lin, r_value_lin, _, _ = linregress(x_original, y_original)
y_pred_linear = slope_lin * x_original + intercept_lin

# Mod√®le exponentiel ajust√© (sur les donn√©es originales)
y_pred_exponential = A_exp * np.exp(B_exp * x_original)

plt.figure(figsize=(12, 7))
plt.plot(x_original, y_original, 'o', label='Donn√©es brutes de concentration')
plt.plot(x_original, y_pred_linear, 'b-', label=f'Ajustement lin√©aire: Y = {slope_lin:.2f}X + {intercept_lin:.2f} (R¬≤ = {r_value_lin**2:.2f})')
plt.plot(x_original, y_pred_exponential, 'g--', label=f'Ajustement exponentiel: Y = {A_exp:.2f}e^({B_exp:.2f}X) (R¬≤_log = {r_value_exp**2:.2f})')
plt.title(f'Comparaison des ajustements lin√©aire et exponentiel ({file_path.split("/")[-1]})')
plt.xlabel(f'{time_column} (unit√© de temps)')
plt.ylabel(f'{concentration_column} (unit√© de concentration)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()

print(f"\nLe graphique de comparaison des deux mod√®les (lin√©aire et exponentiel) sur les donn√©es originales a √©t√© g√©n√©r√©.")
print(f"Observez quelle courbe semble le mieux √©pouser la forme de vos points de donn√©es.")

# Comparaison des R¬≤
print(f"\nComparaison des coefficients de d√©termination (R¬≤) :")
print(f"  R¬≤ pour le mod√®le lin√©aire direct : {r_value_lin**2:.4f}")
print(f"  R¬≤ pour le mod√®le exponentiel (bas√© sur la r√©gression de ln(Y)) : {r_value_exp**2:.4f}")

print(f"\n**Interpr√©tation :**")
print(f"Si le R¬≤ du mod√®le exponentiel (calcul√© sur les donn√©es transform√©es) est significativement plus √©lev√©,")
print(f"et que les r√©sidus de la r√©gression de ln(Concentration) sont plus al√©atoires autour de z√©ro que ceux de la r√©gression lin√©aire directe,")
print(f"alors le mod√®le exponentiel est probablement une meilleure repr√©sentation de la relation sous-jacente.")
print(f"Visuellement, le graphique de comparaison devrait √©galement vous donner une bonne indication.")
````
