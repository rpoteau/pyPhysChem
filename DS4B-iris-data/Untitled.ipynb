```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression

# === Charger le dataset iris ===
iris = load_iris(as_frame=True)
df = iris.frame

# Variables
X = df[["petal length (cm)"]]   # variable explicative
y = df["sepal length (cm)"]     # variable à prédire

# === Régression linéaire ===
model = LinearRegression()
model.fit(X, y)

# === Prédiction ===
y_pred = model.predict(X)

# === Résumé ===
print(f"Coefficient : {model.coef_[0]:.3f}")
print(f"Ordonnée à l'origine : {model.intercept_:.3f}")
print(f"R² = {model.score(X, y):.3f}")

# === Visualisation ===
plt.figure(figsize=(6,4))
plt.scatter(X, y, color="blue", label="Données réelles")
plt.plot(X, y_pred, color="red", label="Régression linéaire")
plt.xlabel("Longueur de pétale (cm)")
plt.ylabel("Longueur de sépale (cm)")
plt.legend()
plt.tight_layout()
plt.show()
```
