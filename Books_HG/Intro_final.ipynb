# Data and tasks

When asking a computer to build a model (= to learn) from data, many different approaches are proposed, but three of them clearly emerge :

- The data you give are including properties and results, and the aim is deduction of the results for a given set of properties: such a problem is called supervised learning, as the data are labelled, that is the expected output is known during the training.  

- The data you give include a number of properties and the aim is to deduce trends from these properties and then apply such trends to a new set of properties: this is unsupervised learning.  

- An alternative approach is called reinforcement learning: the aim is to build the dataset "on the fly", starting from a minimal set of data. The program will probe potential solutions (thanks to an interface with outside the programm) which are evaluated and depending on the results of the evalutation (positive / negative), the model is adapted (reward / penalty).

# Two approaches for a single proble : linear regression

A way to illustrate the difference between inductive and deductive approaches of problems is to examine the regression problem. You have already seen that it is possible to correlate petal width and petal length using a regression. This is a deductive approach : it is stated that a linear regression should work (this is the model) and the best linear regression is computed on the observations thanks to that initially stated model.

```python
import numpy as np
from scipy.stats import linregress
```

As already seen, it is possible link petal length and petal width by a linear correlation as done below : this is a **deductive approach**. Looking at the observations (petal length and width), a model is proposed : there is a linear relationship between these two values. An analytical solution is found to evaluate this relationship and this solution is implemented in a code (here, the scipy linregress commmand) which will calculate the best fit for the model. 

```python
dfi=pd.read_csv('iris.csv', sep=";")
pw=dfi["petal_width"].to_numpy()
pl=dfi["petal_length"].to_numpy()
fit = linregress(x=pw, y=pl)
```

```python
a = fit.slope
b = fit.intercept
print("\nLinear Regression Results:")
print(f"y = {a:.2f} * x + {b:.2f}")
def pl_hat(a,b,x):
    return a * x + b
    
plt.plot(pw, pl,linestyle='', marker='o', color='blue', label='observations ')
plt.plot(pw,pl_hat(a,b,pw),color='red',label='prédiction')
plt.xlabel("petal width") # label de l'axe des abscisses
plt.ylabel("petal length") # label de l'axe des ordonnées
plt.legend() #affichage de la légende sur le graphe
plt.show()
```

An **inductive approach** to the same problem is to consider  starting only with the observations (petal length and width). A programm is then implemented to learn from these data to find the best fit. The more "intelligent" the programm, the best the fit will be.  
The first programm that was proposed to mimick "learning from data" is the **Percpetron** (nvented by Rosenblatt as soon as 1958 !). The principle is summarized in the drawing below :
<p style="text-align: center"><img width="350px" src="../DS4B-svg/1neuron-perceptron-wBackP.png" style="margin-left:auto; margin-right:auto" id="img_perceptron"/></p>
As for the linear regression, the predicted $\hat{y}$ is a linear function of x, but the slope and intercept are not deducted from an analytical model, but are derived thanks to the retroaction in order to minize the difference between $\hat{y}$ and $y$ on the observed data.  

Let us examine the resulting workflow. First, initial values are defined for w and b. Then the amplitude of the backpropagation is chosen (learning_rate) and the number of time the backpropagation is done is defined (epochs : this term will be discussed later).  

```python
# Initialize parameters
weight = 0.0
bias = 0.0

# Hyperparameters
learning_rate = 0.0001
epochs = 100

# History
hist_weight = []
hist_bias = []
```

Then, for all values of the dataset, the $\hat{y}$  is computed for the cuyrrent values of weight and bias. The "distance"  between $\hat{y}$ and the effective value of $y$ is evaluated (loss : various possibilities can be proposed, below it is the square of the difference that is used) and the derivative (gradients) of this loss are computed. Indeed, the aim of the programm is to minimize the loss and various way to minimize can be proposed. Here the one chosen is a gradient descent, that is the backpropagation will be done in the direction of the gradient of the loss : this is done in the "Update weight and bias " step. Finally, for transking purposed, an history of the weight and bias is savec and printed. 

```python
for epoch in range(epochs):
    total_loss = 0
    for x, y_true in zip(pw, pl):
        # Forward pass
        y_pred = weight * x + bias
        
        # Compute loss (Mean Squared Error)
        loss = (y_pred - y_true) ** 2
        total_loss += loss
        
        # Compute gradients
        dL_dy = 2 * (y_pred - y_true)
        dy_dw = x
        dy_db = 1
        
        # Update weights and bias
        weight -= learning_rate * dL_dy * dy_dw
        bias   -= learning_rate * dL_dy * dy_db
        
    if epoch % 10 == 0 or epoch == epochs - 1:
        hist_weight.append(weight)
        hist_bias.append(bias)
        avg_loss = total_loss / len(pw)
        print(f"Epoch {epoch}, Loss: {avg_loss:.4f}, Weight: {weight:.4f}, Bias: {bias:.4f}")

# Final model
print("\nTrained neuron:")
print(f"y = {weight:.2f} * x + {bias:.2f}")
```

The obtained weight and bias are mostly similar to the values obtained from the linear regression: the programm has learnt the values of wieght and bias instead of getting them analytically.  

```python
# Visualization
x_vals = list(range(0, 4))
y_vals = [weight * x + bias for x in x_vals]
hist_y_vals = []

for i in range(len(hist_weight)-1):    
    temp=[hist_weight[i] * x + hist_bias[i] for x in x_vals]
    hist_y_vals.append(temp)
   
plt.figure(figsize=(8, 5))
for i in range(len(hist_weight)-1):
   plt.plot(x_vals, hist_y_vals[i], color='green')
plt.scatter(pw, pl, color='red', label='Training Data')
plt.plot(x_vals, y_vals, color='black', label='Trained Neuron Output')
plt.xlabel('Input (x)')
plt.ylabel('Output (y)')
plt.title('Single Neuron Fit')
plt.legend()
plt.grid(True)
plt.show()
```

## Preparing the data   
The first step of a machine learning project is to find a way to validate the results obtained. Just as a student needs to pass test to confirm he has properly learnt his lessons, a machine learning model needs to be tested to validate the model. This is done by splitting the database in two part: one used for training, and one used for testing. 

```python
data_train = dfi.sample(frac=0.8, axis='index') 
data_test  = dfi.drop(data_train.index)
x_train = data_train[['petal_width']]
y_train = data_train[['petal_length']]
x_test = data_test[['petal_width']]
y_test = data_test[['petal_length']]
print("composition du training set", data_train['species'].value_counts(normalize=True))
print("composition du test set", data_test['species'].value_counts(normalize=True))
```

By ré-exécuting many times this cell, you see that the splitting is not always done the same way, and that the three species, which where equally represented in the initial dataset, are no more in the training set or the test set. In some cases, this might play a major role. In the study ["Predicting reaction perfomance in C-N cross-coupling using machine learning"](https://www.science.org/doi/10.1126/science.aar5169), the authors wants to study reactivity of this reaction as function of the substitution of the two reactants, the catalyst, the base and the additives used.   
When trying to predict yield, a special attention has to be given when splitting the dataset so that a sample of each partner is introduced in the dataset. In this way, the training is carried out on the whole chemical space and the possibility to predict new coupling with the same chemical space can be evaluated with the test set.  
In contrast, when trying to predict the reactivity of a new additive from the results obtained with other, the splitting of the dataset is done with all the results of some additives in the train set, and all the results of other additives in the test set. 
In our case, if we want to predict pl from pw for all kind of species, it is important to keep balance between the species in both sets. 
There is a variety of solution to split the dataset in a "smart" way. Here is below a solution (proposed by mistral AI) using only panda, but others exist either based on keras or on [scikit-learn](https://scikit-learn.org/), a more simple python library for data analysis. 

```python
# Taille du test set (20%)
test_size = 0.2

# Initialiser des listes vides pour stocker les indices
train_indices = []
test_indices = []

# Pour chaque classe, échantillonner de manière stratifiée
for _, group in dfi.groupby('species'):
    n_test = int(len(group) * test_size)
    test_sample = group.sample(n=n_test, random_state=42)
    train_sample = group.drop(test_sample.index)

    test_indices.extend(test_sample.index)
    train_indices.extend(train_sample.index)

# Créer les training et test sets
train_set = dfi.loc[train_indices]
test_set = dfi.loc[test_indices]

# Vérification de la répartition des classes
print("Répartition dans le training set :\n", train_set['species'].value_counts(normalize=True))
print("Répartition dans le test set :\n", test_set['species'].value_counts(normalize=True))
```

## Structure and training the model  
We then need to describe how the ANN is built. As we want to work with a perceptron, we will use the **sequential model API**, that is the simplest API. It involves defining a [Sequential class](https://keras.io/api/models/sequential/), that is adding layers to the model one by one in a linear manner, from input to output. The perceptron is a Sequential MLP model that accepts one input and that has an output layer with one node to predict a numerical value.  
The compile part defines the way retroaction is carried out: the loss function is the way the distance between predicted and real data is evaluates (below mse), the optimizer is the way the gradient is used to go down 

```python
def define_1ANN():
    
    model = keras.models.Sequential()
    model.add(keras.layers.Input(shape=(1,), name='inpLayer'))
    model.add(keras.layers.Dense(1, name='outLayer'))
    
    model.compile(optimizer = 'sgd',
                  loss      = 'mse',
                 )
    return model

ANNmodel=define_1ANN()
ANNmodel.summary()
```

We now need to make the training. We will use exactly the same parameters as used before in the case of our model perceptron : 100 epochs and we will get the weight and bias. It is time to introduce the mention of batch. In the model perceptron used before, the weight and biase are modified after calculation of mse for each sample. But the mse could have been sumed over a serie of sample, and retroaction could have been made after this serie of sample. Such a serie of sample is called a batch and in the following, the batch_size has been set to one to reproduce what has been done in the case of our model perceptron. Various solutions can be proposed for batch_size:  
- stochastic gradient descent : where the batch size is one and when you make retroaction after each sample
- batch gradient descent : where you sample all the training set before making retroaction and the batch_size is the size of the training set.
- mini-batch gradient descent : where you split your training set in many batches (which are different for each epochs and self generated by the model). 



```python
ANNhistory = ANNmodel.fit(x_train,
                    y_train,
                    epochs          = 100,
                    batch_size      = 1,
                    verbose         = 1)
```

```python
weights = ANNmodel.layers[0].get_weights()  # Accès aux poids de la première couche Dense
print(weights)
```

However, at that point, we have no way to know if our parameters are the best: could we use less epochs ? or should we use more ? Do answer this question we need to examine the history of the training process. This is done below. First, we observe that most of the low convergence is obtained in a few epochs (< 20). Then, we observe a quite noisy loss: this could be tuned by modifying batch_size or using a more efficient optimizer. Second, we only have history on the training set: it is also important to follow the results on the test set. 

```python
dfh=pd.DataFrame(data=ANNhistory.history)
figLOSS=dfh.plot(y=["loss"],linestyle='-', marker='o',fontsize=14)
figLOSS.set_xlabel('epoch',fontdict={'fontsize':16})
figLOSS.set_ylabel('loss',fontdict={'fontsize':16})
figLOSS.set_ylim([0.0,0.5])
figLOSS.legend(loc='upper right', shadow=True, fontsize='x-large')
figLOSS.axhline(y=0.2130046784877777, color='r', linestyle='--', linewidth=1, label='Asymptote')
```

Let us thus try with new paramters. in the definition of the model, we will change to a more complex optimizer and we add more metrics to follow : MAE is added which is more comparable directly to the values studied. 

```python
def define_1ANN_best():
    
    model = keras.models.Sequential()
    model.add(keras.layers.Input(shape=(1,), name='inpLayer'))
    model.add(keras.layers.Dense(1, name='outLayer'))
    
    model.compile(optimizer = 'sgd',
                  loss      = 'mse',
                  metrics   = ['mae', 'mse'] )
    return model
best_ANNmodel=define_1ANN_best()
best_ANNmodel.summary()
```

Then this model is run, but a **callback** is added with the command `EarlyStopping` callback stops training when a monitored metric has stopped improving. Check if early stopping really happens. If not, it means the model is not converged : the model should be run a second (or more) time until it converges. As previously noticed, batch_size in added to smooth the optimization process. And finally, validation data are defined for which the metrics will also be computed. 

```python
from tensorflow.keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=10)
best_ANNhistory = best_ANNmodel.fit(x_train,
                    y_train,
                    epochs          = 100,
                    batch_size      = 10,
                    verbose         = 1,
                    validation_data = (x_test, y_test),
                    callbacks=[es] )

```

When evaluating the models, you now have three metrics : loss (mse), and then mae and mse again as requested. 

```python
best_score_train = best_ANNmodel.evaluate(x_train, y_train, verbose=0)
best_score_test = best_ANNmodel.evaluate(x_test, y_test, verbose=0)
print("score training set", best_score_train, "score test set", best_score_test)
```

Plotting the evaluation functions (loss, mae or mse) as a function of epochs is necessary, not only to evaluate model convergence (the loss is supposed to decrease over time), but also to check for overfitting and underfitting:
- when the model learns the training data too well, including its noise and outliers, and fails to generalize to new, unseen data,  the loss on the training data continues to decrease while the loss on a test (validating) set starts increasing
- when the loss is still high after many epochs, the model is too simple to capture the underlying patterns in the data : this is underfitting.

```python
dfh=pd.DataFrame(data=best_ANNhistory.history)
figLOSS=dfh.plot(y=["loss","val_loss"],linestyle='-', marker='o',fontsize=14)
figLOSS.set_xlabel('epoch',fontdict={'fontsize':16})
figLOSS.set_ylabel('loss',fontdict={'fontsize':16})
figLOSS.set_ylim([0.0,0.5])
figLOSS.legend(loc='upper right', shadow=True, fontsize='x-large')
figLOSS.axhline(y=0.2130046784877777, color='r', linestyle='--', linewidth=1, label='Asymptote')
```
